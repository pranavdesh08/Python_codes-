## What is Pandas?

### Pandas is a library used in Python programming language. It is used to structure the data in a table format. The data is stored in 1-D/2-D format which is divided into number of rows and columns. Pandas is also used to read a particular file, it can be in csv, json, txt, and many other files with audio(wave). Pandas is built on Numpy, it has similarities with Numpy such as in creating a DataFrame and indexing. In order to use Pandas we have to install it in our system.

```python
#import pandas library
import pandas as pd
```

```python
# Creating a DataFrame using list
make_1=[['cool',1],['hot',88]]
make_1_df=pd.DataFrame(make_1,columns= ['condition', 'numbers'])
make_1_df
```

```python
#DataFrame using Dictionary
make_2={'condition':[45,22,66,88],'numbers':[99,24,56,12]}
make_2_df=pd.DataFrame(make_2)
make_2_df
```

```python
#Changing the index of the table.
make_3={'condition':[45,22,66,88],'numbers':[99,24,56,12]}
make_3_df=pd.DataFrame(make_3,index=['a','b','c','d'])
make_3_df
                       
```

```python
# Create DataFrame from list of dictionary
make_4=[{'India':1,'USA':5,'Aus':5},{'India':55,'USA':20,'Aus':77}]
make_4_df=pd.DataFrame(make_4)
make_4_df
```

```python
#Changing index
make_4=[{'India':1,'USA':5,'Aus':5},{'India':55,'USA':20,'Aus':77}]
make_4_df=pd.DataFrame(make_4,index=['a','b'])
make_4_df
```

```python
# DataFrame index.
make_4=[{'India':1,'USA':5,'Aus':5},{'India':55,'USA':20,'Aus':77}]
make_4_df=pd.DataFrame(make_4,index=['a','b'],columns=['Aus','India'])
make_4_df
```

```python
# Change Columns name.
make_4_df.columns=['a1','b1']
make_4_df
```

```python
# Location function 
make_3_df
```
